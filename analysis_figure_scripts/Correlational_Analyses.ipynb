```python
import pandas as pd
import numpy as np
import seaborn as sns
import EmotionEvaluationCorrelationHelper as helper
import matplotlib.pyplot as plt
import matplotlib
from IPython.core.pylabtools import figsize
from matplotlib.colors import ListedColormap
import scipy.stats as stats
from labellines import labelLine, labelLines
from matplotlib.patches import Rectangle
import scipy.stats as stats
```

```python
df_study_1 = pd.read_csv('../data/study1_olr.csv', delimiter='\t')
df_study_2 = pd.read_csv('../data/study2_olr.csv', delimiter='\t')
df_study_3 = pd.read_csv('../data/study3_olr.csv', delimiter='\t')
df_study_4 = pd.read_csv('../data/study4_olr_formatted.csv', delimiter=',')
```

```python
def putRValuesOnDiagonal(corr_df,z_scores):
    for emotion in corr_df.emotions:
        r_val = corr_df.loc[corr_df['emotions']==emotion]['corr_val_dim'].values[0]
        z_scores[emotion][emotion] = r_val
    return z_scores

def BoldRValues(ax, z_scores):
    texts = ax.texts
    for i in range((len(texts)-len(z_scores)),len(texts)):
        text = texts[i]
        text.set_weight('bold')

def graphZscoreHeatMap(z_scores,
                        bonferroni_zscore,
                        title,
                        corr_df):
    figsize(15,15)
    plt.rcParams['font.family']="Arial" 
    plt.rcParams['font.size']="12" 
    fmt_string ='.2f'
    heat = sns.heatmap(z_scores, 
                       mask=z_scores < 0,
                       cmap=ListedColormap(['white']), 
                       annot=True,
                       fmt=fmt_string,
                       cbar=False)
    heat = sns.heatmap(z_scores, 
                       mask=z_scores < 2.3263,
                       cmap=ListedColormap(['grey']),
                       annot=True,
                       fmt=fmt_string,
                       cbar=False)
    heat = sns.heatmap(z_scores, 
                       mask=z_scores < bonferroni_zscore,
                       cmap=ListedColormap(['black']), 
                       fmt=fmt_string,
                       annot=True,
                       cbar=False)
    
    empty_df=z_scores
    empty_df[:][:]= np.NAN
    r_val_on_diagonal = putRValuesOnDiagonal(corr_df,empty_df)
    r_val_map = sns.heatmap(r_val_on_diagonal, 
                            cmap=ListedColormap(['white']),
                            alpha=0.0,
                            annot=True, 
                            cbar=False,
                            fmt=fmt_string)
    heat.set_title(title)
    BoldRValues(r_val_map, z_scores)
    plt.yticks(rotation=0)
    plt.xticks(rotation=90)
    ax = heat.axes
    for i in range(len(corr_df)):
        ax.add_patch(Rectangle((i, i), 1, 1, fill=False, edgecolor='black', lw=2))
    return heat

def GetRowStartIndex(row_num):
    index = 0
    if(row_num>0):
        index= row_num + GetRowStartIndex(row_num-1)
    return index

def createZscoreHeatMap(df,emo_val_dim_1, emo_val_dim_2):
    
    corr_df = helper.getPairwiseCorrelations(df,emo_val_dim_2,emo_val_dim_1)
    sorted_corrs = corr_df.sort_values('corr_val_dim',ascending=True)
    sorted_corrs = sorted_corrs.reset_index(drop=True)
    sorted_emotions = list(sorted_corrs['emotions'])
    
    z_scores = helper.createZScoreDf(df, 
                                      sorted_emotions,
                                      emo_val_dim_2,
                                      emo_val_dim_1)
    num_rows = len(z_scores)
    bonferroni_p = .05/(num_rows*(num_rows-1)/2)
    z_score_black_limit = abs(stats.norm.ppf(bonferroni_p))
    heat = graphZscoreHeatMap(z_scores,z_score_black_limit,"",sorted_corrs)
    
```
